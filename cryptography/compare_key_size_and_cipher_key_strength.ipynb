```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = [8, 5]
plt.rcParams["figure.autolayout"] = True
```

```python
x=[ i*i for i in range(2,128)]
y1 = [1/(2^t) for t in x]
y2 = [1-u for u in y1]
```

```python
plt.plot(x, y1, color='g', label='p')
plt.plot(x, y2, color='r', label='q=1-p')
plt.xlabel("Number of Bits Used to represent the Key")
plt.ylabel("Probability")
plt.title("Plotting the probabiity of guessing and not guessing the right key for One-Time Pad Cipher")
plt.legend()
plt.show()
```
